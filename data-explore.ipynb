```python
import pandas as pd
import app_data
import folium
from geopy import MapBox
from geopy.extra.rate_limiter import RateLimiter
import plotly.express as px
```

```python
df = pd.read_csv("customerdb.csv")
```

```python
#strip strings
for col in ['Chain', 'TABC Permit']:
    df[col] = df[col].str.strip()
```

```python
#filter out grocery chains
chain_list = ["Kroger", "HEB", "Whole Foods", "Central Market", "Martket Street", "Albertsons", 'Fresh Market','Target', 'Tom Thumb', 'Walmart', "Trader Joe's"]
df = df[df["Chain"].isin(chain_list)]
```

```python
df.loc[df.duplicated('TABC Permit', keep=False), ('TABC Permit', 'CUSTOMER', 'Street')].sort_values('TABC Permit')
```

```python
df['TABC Permit'] = df['TABC Permit'].str.replace(" ","")
df = df.dropna(subset=['TABC Permit'])
df = df.drop_duplicates('TABC Permit', keep=False)
df['TABC Permit'].sort_values().unique()
```
